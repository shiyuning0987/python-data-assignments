```python
import bs4
import requests
from bs4 import BeautifulSoup 
import csv
```

```python
#获取考拉美妆类一级页面的第一页
r = requests.get('https://www.kaola.com/category/1473.html?key=&pageSize=60&pageNo=1&sortfield=2&isStock=false&isSelfProduct=false&isPromote=false&isTaxFree=false&factoryStoreTag=-1&isDesc=true&b=&proIds=&source=false&country=&needBrandDirect=false&isNavigation=0&lowerPrice=-1&upperPrice=-1&backCategory=&headCategoryId=&changeContent=type&#topTab')
r.encoding = 'utf-8'
mypage = BeautifulSoup(r.text,"html.parser") 
goods=mypage.find('div', class_='m-result')
goods
```

```python
#建立循环，拉取页面变量
dataset= []
goodslist=mypage.find_all('li',class_='goods colorsku')
for myitem in goodslist:
    name = myitem.find('h2').text.strip()
    curprice =myitem.find('span', class_='cur').text.strip()
    country =myitem.find('span', class_='proPlace').text.strip()
    store =myitem.find('p', class_='selfflag').text.strip()
    sales =myitem.find('a', class_='comments').text.strip()
    dataset.append([name,curprice,sales,country,store])
dataset
```

```python
#打开考拉美妆类第二页
r2= requests.get('https://www.kaola.com/category/1473.html?key=&pageSize=60&pageNo=2&sortfield=2&isStock=false&isSelfProduct=false&isPromote=false&isTaxFree=false&factoryStoreTag=-1&isDesc=true&b=&proIds=&source=false&country=&needBrandDirect=false&isNavigation=0&lowerPrice=-1&upperPrice=-1&backCategory=&headCategoryId=&#topTab')
r2.encoding = 'utf-8'
mypage2 = BeautifulSoup(r2.text,"html.parser") 
goods2=mypage2.find('div', class_='m-result')
goods2
```

```python
#重复循环拉取变量
dataset2 = []
goodslist2=mypage2.find_all('li',class_='goods colorsku')
for myitem2 in goodslist2:
    name2 = myitem2.find('h2').text.strip()
    curprice2 =myitem2.find('span', class_='cur').text.strip()
    country2 =myitem2.find('span', class_='proPlace').text.strip()
    store2 =myitem2.find('p', class_='selfflag').text.strip()
    sales2 =myitem2.find('a', class_='comments').text.strip()
    dataset2.append([name2,curprice2,sales2,country2,store2])
    
print(dataset2)
```

```python
#组合两个列表，写入表格
final=dataset+dataset2
with open('kaola2.csv','w',newline='',encoding='utf8') as f:
    writer = csv.writer(f)
    header = ['Name','Price','Comments','Country','Store']
    writer.writerow(header)
    writer.writerows(final)
```
